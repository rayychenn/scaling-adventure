```python
def return_clean_df(pathway):
    df = pd.read_parquet(pathway)
    df['transaction_hash'] = df['transaction_hash'].apply(lambda x: '0x' + x.hex())
    df['from_address'] = df['from_address'].apply(lambda x: '0x' + x.hex())
    df['to_address'] = df['to_address'].apply(lambda x: '0x' + x.hex() if x is not None else x)
    df['block_number'] = df['block_number'].astype(int)
    df['input'] = df['input'].apply(lambda x: '0x' + x.hex() if x  != b'' else '')
    df.drop(['transaction_index', 'max_priority_fee_per_gas', 'max_fee_per_gas', 'chain_id'], axis=1, inplace=True)
    
    # This is likely just a normal transfer.
    if 'value' in df.columns:
        df = df[~((df['input'] == '') & (df['value'] != 0))]
    else:
        df = df[~((df['input'] == '') & (df['value_string'] != 0))]        
        df['value'] = df['value_string']
        del df['value_string']
    maxi, mini = df['block_number'].max(), df['block_number'].min()
    df = pd.merge(df, block_timestamps[block_timestamps['number'].between(mini, maxi)], left_on='block_number', right_on='number')
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df['week'] = df['timestamp'].dt.week
    df['hour'] = df['timestamp'].dt.hour
    df['year'] = df['timestamp'].dt.year
    return(df)

# Split the contents based on the weeks.

def return_week_split(pathway):
    df = return_clean_df(pathway)
    if len(df) == 0:
        return({})
    week_dict = {}
    start_time = df.iloc[0]['timestamp'].week
    end_time = df.iloc[-1]['timestamp'].week
    if start_time != end_time:
        for week in [start_time, end_time]:
            week_dict[week] = df[(df['week'] == week)].copy()
    else:
        week_dict[start_time] = df.copy()
    
    return(week_dict)
        
```
