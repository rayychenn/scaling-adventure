```python
def identify_top_addresses_for_the_week(file_path, minimum_number=500, maximum_return=10):
    
    df = pd.read_parquet(file_path, engine='fastparquet', columns=['to_address', 'from_address'])
    unique_interactions = df[['from_address', 'to_address']].drop_duplicates().groupby('to_address').agg({'from_address':'count'})
    unique_interactions = unique_interactions[unique_interactions['from_address'] > minimum_number].sort_values('from_address', ascending=False).head(maximum_return)
    return(unique_interactions.index.to_list())
```

```python
all_addrs = {}

for file_path in sorted([f"cleaned/{i}" for i in os.listdir('cleaned') if '.parquet' in i])[:-2]:
    all_addrs[file_path]= identify_top_addresses_for_the_week(file_path)
    print(file_path)
```

```python
min_appearance = 8
filtered_addrs = []
for j in all_addrs.values():
    filtered_addrs += j[:20]

common_addrs = pd.DataFrame([dict(collections.Counter(filtered_addrs))])
common_addrs = common_addrs.T.sort_values(0, ascending=False)
common_addrs = common_addrs[common_addrs[0] >= min_appearance]
final_count = common_addrs.index.to_list()
```
